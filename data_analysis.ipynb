# Data Analysis

This notebook contains all code for analysing the generated data.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Import the data
plane = pd.read_csv('Data/plane.csv', header=None)
cylinder = pd.read_csv('Data/cylinder.csv', header=None)
torus = pd.read_csv('Data/torus.csv', header=None)

# Assign the column names
columns = ['Closed curves', 'Open curves', 'Total curves', 'Water components', 'Land components', 'Total components', 'Mask']
plane.columns = columns
cylinder.columns = columns
torus.columns = columns
```

## Plane data analysis

```python
display(plane.describe())
```

```python
# Custom bins
bin_edges = np.arange(plane["Total curves"].min() - 0.5, plane["Total curves"].max() + 1.5, 1)

# Plot the distribution of "Total curves"
plt.figure(figsize=(8, 6))
counts, bins, bars = plt.hist(plane["Total curves"], bins=bin_edges, edgecolor="k", alpha=0.7, density=True)  # Normalize histogram

# Annotate each bar with the value
for count, bar in zip(counts, bars):
    if count > 0:  # Annotate only non-zero heights
        if count > 0.0001:
            plt.text(bar.get_x() + bar.get_width() / 2, count, f"{count:.4f}", ha="center", va="bottom", fontsize=9, color="black")
        else:
            plt.text(bar.get_x() + bar.get_width() / 2, count, f"{count:.2e}", ha="center", va="bottom", fontsize=9, color="black")

# Customize the plot
plt.title("Normalized Distribution of Total Curves in Plane map")
plt.xlabel("Total Curves")
plt.ylabel("Density")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.xticks(bin_edges + 0.5)  # Center the tick marks
plt.show()
```
